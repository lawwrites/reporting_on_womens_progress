```python
# Obtaining data for plotting
male = wdf_s[(wdf_s['measure'] == 'Time spent on unpaid care and domestic work') & (wdf_s['sex'] == 'Male')]
female = wdf_s[(wdf_s['measure'] == 'Time spent on unpaid care and domestic work') & (wdf_s['sex'] == 'Female') & ~(wdf_s['unit_of_measure'] == "Factor of men's time")]

#merge the datasets on 'ref_code' to ensure matching countries
merged_df = pd.merge(male[['ref_code', 'obs_value']], female[['ref_code','obs_value']], on='ref_code', suffixes=('_male','_female'))


# Group by countries or categories
x = np.arange(len(merged_df['ref_code']))  # the label locations

#Ensure that male and female datasets have matching lengths and are sorted consistently
assert len(male) == len(female), "Male and female datasets must have same length."

# Setting up the bar chart parameters
width = 0.6  # Adjust the bar width
fig, ax = plt.subplots(figsize=(12, 6))  # Create the figure and the axis

# Plotting bars
p1 = ax.bar(x, merged_df['obs_value_male'], width, label='Male', color=palette[1])
p2 = ax.bar(x, merged_df['obs_value_female'], width, label='Female', bottom=merged_df['obs_value_male'], color=palette[7])

# Setting the x-tick labels based on your categories (e.g., 'ref_code')
ax.set_xticks(x)
ax.set_xticklabels(male['ref_code'], rotation=45, ha='right')
ax.set_yticks(np.arange(0,11,1))

# Show titles, labels, and legend
ax.set_title("Time spent on unpaid care and domestic work (male vs. female)", title_font)
ax.set_xlabel("Countries", fontdict=label_font)
ax.set_ylabel("Observation Value", fontdict=label_font)
ax.legend(bbox_to_anchor=(1.15,1), loc="upper right")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

## Findings:

#### Scatterplot and Stacked Bar Plot
> In all 12 countries, males spend significantly less time on unpaid care and domestic work compared to females. In many cases, males spend less than half the time that females do on household responsibilities. Mexico and India show the largest gap between males and females while Canada, the USA, and the United Kingdom have the smallest gaps.

# Multivariate Exploration

```python
#Create dataframe to focus on three catagories: reproductive autonomy, access to financial assets, and household responsibilities
sg_df = wdf_s[(wdf_s['measure'] == 'Gender gap in managerial positions') | (wdf_s['measure'] == 'Perception that being a housewife is as fulfilling as working for pay') | (wdf_s['measure'] == 'Legal discrimination on reproductive autonomy')]

#Reset index
sg_df.reset_index(inplace=True,drop=True)
#Sort values by reference area
sg_df= sg_df.sort_values(by='ref_code')
```

```python
#Verify the dataframe worked
print(sg_df.info())
#Print it out to check it
sg_df.head(1)
```

## Prediction for comparing different gender-related metrics: 
### Observation scores will show a relationship
> Perception of the role of the housewife being more fulfilling than working for pay as well as legal discrimination on reproductive autonomy will impact gender gaps in the workplace.

```python
# Create the markers for each sex type
measure_markers = [['Legal discrimination on reproductive autonomy', 'D', palette[0]], ['Gender gap in managerial positions', 'o', palette[4]],
                   ['Perception that being a housewife is as fulfilling as working for pay', 's', palette[6]]]

# Ensure ref_code is a string
sg_df['ref_code'] = sg_df['ref_code'].astype(str)


# Loop through measure types and plot each one
for measure, marker, color in measure_markers:
    # Filter data by 'measure' column
    plot_data = sg_df.loc[sg_df['measure'] == measure]
    
    # Check if plot_data is not empty
    if not plot_data.empty:
        sns.scatterplot(data=plot_data, x='ref_code', y='obs_value', marker=marker, label=measure, color=color)

#Choosing colors
sns.color_palette("husl",3)
# Add titles and labels
plt.xlabel('Countries', label_font)
plt.ylabel('Observation Values', label_font)
plt.title('Comparing different gender-related metrics', title_font)

# Show legend
plt.legend(bbox_to_anchor=(1.05,1), loc="upper left", borderaxespad=0)

# Create figure and axis
plt.figure(figsize=(15, 8))
plt.tight_layout()

# Show plot without scrolling bars
plt.show()
```

```python
# Obtaining the filtered data
hw = wdf_s[wdf_s['measure'] == 'Perception that being a housewife is as fulfilling as working for pay']
gg = wdf_s[wdf_s['measure'] == 'Gender gap in managerial positions']
ra = wdf_s[wdf_s['measure'] == 'Legal discrimination on reproductive autonomy']

# Setting up the bar chart parameters
x = np.arange(len(hw['ref_code']))  # the label locations based on the length of 'hw'
width = 0.25  # The width of the bars
multiplier = 0

# Creating the plot and axes
fig, ax = plt.subplots(figsize=(14, 7), layout='constrained')  # layout to prevent overlapping labels

# Looping through measurement and plotting bars
for data, label, color in zip([hw, gg, ra], ['Perception of housewife is as fulfilling as working for pay', 'Gender Gap in management', 'Legal discrimination in reproductive autonomy'], [palette[1], palette[2],palette[3]]):
    offset = width * multiplier
    rects = ax.bar(x + offset, data['obs_value'], width, label=label, color=color)
    multiplier += 1

# Set axis labels and title
ax.set_xlabel('Countries', fontdict=label_font)
ax.set_ylabel('Observation Value', fontdict=label_font)
ax.set_title('Comparing Different Gender-Related Metrics Across Countries', fontdict=title_font)

# Set x-tick labels
ax.set_xticks(x + width)  # Adjust x-ticks to align with bar groups
ax.set_xticklabels(hw['ref_code'], rotation=45, ha='right')

# Adding a legend
ax.legend(bbox_to_anchor=(1.03,1), loc="upper left", borderaxespad=0)

# Show the plot
plt.show()
```

## Findings:
> When examining the gender gap in managerial positions, countries are progressing towards zero discrimination, with all 12 countries having an observation value of 40 or less. However, this progress does not necessarily indicate improvement in legal rights for reproductive autonomy or in valuing the role of housewives as paid work. Therefore, there is no clear relationship between these categories.

### Were there any interesting or surprising interactions between features?

> It is surprising to find that while women are closing the professional gap in countries with the world's largest GDPs, they still struggle with gaining reproductive autonomy and shifting public perceptions to empower working wives (with or without children). There appears to be a disconnect between women's increased participation in leadership roles and their autonomy over reproductive rights, as well as the societal value placed on the work they do as wives and mothers.
